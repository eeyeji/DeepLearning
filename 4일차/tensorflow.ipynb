```python
import numpy as np
```

```python
#y=2x+1 관계를 만족하는 트레이닝에 사용할 데이터를 만든다.
x_train = np.array([0,1])
y_train = x_train*2+1
print("train")
print(x_train)
print(y_train)

x_test = np.array([2,3])
y_test = x_test*2+1
print("test")
print(x_test)
print(y_test)
```

```python
#학습모델 생성
model = keras.models.Sequential()
model.add(keras.layers.Dense(1, input_shape=(1,)))
model.summary()
```

```python
#학습전 결과는 예측값과 결과값이 일치하지 않음
y_predict=model.predict(x_test)
print(y_predict.flatten())
print(y_test)
```

```python
model.compile("SGD", "mse")
```

```python
#model.fit(x_train, y_train, epochs=1000, verbose=0)
model.fit(x_train, y_train, epochs=1000, verbose=1)
```

```python
y_predict = model.predict(x_test)
print(y_predict.flatten())
print(y_test)
```
